##1. How many passengers are in our passenger list?##
From here forward, we’ll assume our dataset represents the full passenger list for the Titanic.

```python
# print df[['PassengerId']].describe() 
print 'There are {0} passengers in the list'.format(df['PassengerId'].count())
```

##2. What is the overall survival rate?##

```python
print 'Overall Survival Rate: {0:.2f} %'.format(df.Survived.mean()*100)
```

##3 & 4. How many male and female passengers were onboard?

```python
# 3
print 'Male: ',df.Sex[df.Sex=='male'].count()

#4
print 'Female: ',df.Sex[df.Sex=='female'].count()
```

##5 & 6. What is the overall survival rate of male and female passengers?

```python
print 'Male Survival Rate: {0:.2f} %'.format(df.Survived[df.Sex=='male'].mean()*100)
print 'Female Survival Rate: {0:.2f} %'.format(df.Survived[df.Sex=='female'].mean()*100)
```

##7. What is the average age of all passengers onboard?

```python
# There are a lot of 'NA' fields in the table. 
# However, if I calculate the average and fill the NAs with the calculated average age,
# the overall average age remains the same

print '{0:.2f}'.format(df.Age.mean())

# At a later time, I can fill the average ages by gender and then calculate the average age 
# to see if it would be different than this
```

##8. What is the average age of passengers who survived?

```python
print '{0:.2f}'.format(df.Age[df.Survived == 1].mean())
```

##9. What is the average age of passengers who did not survive?

```python
print '{0:.2f}'.format(df.Age[df.Survived == 0].mean())
```

##11. How many passengers are in each of the three classes of service (e.g. First, Second, Third?)

```python
# First Class:
print 'First Class:', df.Survived[df.Pclass == 1].count()

# Second Class:
print 'Second Class:', df.Survived[df.Pclass == 2].count()

# Third Class:
print 'Third Class:', df.Survived[df.Pclass == 3].count()
```

##12. What is the survival rate for passengers in each of the three classes of service?

```python
# First Class:
print 'First Class: {0:.2f} %'.format(df.Survived[df.Pclass == 1].mean()*100)

# Second Class:
print 'Second Class: {0:.2f} %'.format(df.Survived[df.Pclass == 2].mean()*100)

# Third Class:
print 'Third Class: {0:.2f} %'.format(df.Survived[df.Pclass == 3].mean()*100)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt 
import numpy as np
import math

# x = Classes
x = sorted(df.Pclass.unique())
y = df.groupby('Pclass').Survived.mean()
print x
print y
```

##13. What else might you conclude?

At a glance, the data shows that:
1. There are more passengers of lower classes (lowest as Third Class) on board than higher (highest as First) classes;
2. Drastic drop of survival rate for lower classes (3rd class has the lowest and 1st class has the highest survival rate)


```python
# Plot Survival Rate versus Classes

fig, ax = plt.subplots(nrows=1, ncols=1)
ax.plot(x, y, color='red')
ax.set_xlabel('Class')
ax.set_ylabel('Survival Rate')
ax.set_title('A simple graph of Survival Rate vs. Classes');
```
