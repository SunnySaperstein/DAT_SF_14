```python
#11 How many passengers are in each of the three classes of service (e.g. First, Second, Third?)

# First Class:
print 'Second Class:' and df.Survived[df.Pclass == 1].count()

# Second Class:
print 'Second Class:' and df.Survived[df.Pclass == 2].count()

# Third Class:
print 'Third Class:' and df.Survived[df.Pclass == 3].count()
```

```python
#12 What is the survival rate for passengers in each of the three classes of service?

# First Class:
print 'First Class:', df.Survived[df.Pclass == 1].mean()

# Second Class:
print 'Second Class:', df.Survived[df.Pclass == 2].mean()

# Third Class:
print 'Third Class:', df.Survived[df.Pclass == 3].mean()
```

```python
#13 What else might you conclude?

"""
At a glance, the data shows that:
1. There are more passengers of lower classes (lowest as Third Class) on board than higher (highest as First) classes;
2. Much less survival rate for lower classes (3rd class has the lowest and 1st class has the highest survival rate)
"""

%matplotlib inline
import matplotlib.pyplot as plt 
import numpy as np
import math

# x = Classes
x = sorted(df.Pclass.unique())
y = df.groupby('Pclass').Survived.mean()
print x
print y
```

```python
# Plot y versus x
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.plot(x, y, color='red')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('A simple graph of Survival Rate vs. Classes');
```
